##1. 기쁨_연인

```python
from google.colab import drive
import pandas as pd
import glob

drive.mount('/content/drive')
```

```python
folder_path = '/content/drive/MyDrive/공감형 대화(부부, 연인)/VS_기쁨_연인'
```

```python
all_files = glob.glob(folder_path + "/*.tsv")
all_data = pd.DataFrame()

for file_path in all_files:
  df = pd.read_csv(file_path, sep='\t')
  all_data = pd.concat([all_data, df])

happy_gb = all_data[['utterance_type', 'utterance_text']]
happy_gb.head()
```

```python
happy_gb.info()
```

```python
all_files = glob.glob(folder_path + "/*.tsv")
happy_total = pd.DataFrame()

for file_path in all_files:
  df = pd.read_csv(file_path, sep='\t')

  happy_text_to = df[df['utterance_type'] == 0][['utterance_text']]
  happy_text_me = df[df['utterance_type'] == 1][['utterance_text']]

  merged_text = pd.concat([happy_text_to.reset_index(drop = True), happy_text_me.reset_index(drop = True)], axis = 1, ignore_index = True)
  happy_total = pd.concat([happy_total, merged_text])

happy_total.head()
```

```python
happy_total.info()
```

```python
happy_total['sentiment'] = '기쁨'
```

```python
happy_total
```

```python
happy_total.to_csv('/content/drive/My Drive/공감형 대화(부부, 연인) (1)/기쁨_연인_통합.csv', encoding='utf-8-sig', index=False)
```
